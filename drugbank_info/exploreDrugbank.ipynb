```python
import pandas as pd
import numpy as np

def parse_list(x):
    # if not stored as an empty list in csv
    if x != '[]':
        return x.strip("['']").split("', '")
    else:
        return []

# Read the CSV file and use the custom parser for the 'categories' column
drugbank_df = pd.read_csv('drugbankConcise.csv', sep=',', converters={'categories': parse_list})
```

```python
drugbank_df.head()
```

```python
null_categories = []
for i in range(0, 12226):
    # checking whether categories list is empty
    if len(drugbank_df['categories'][i]) == 0:
        null_categories.append(i)
        
print('sm w/o SMILES: ', drugbank_df['SMILES'].isna().sum())
print('sm w/o classes: ', drugbank_df['direct_parent'].isna().sum())
print('sm w/o targets: ', drugbank_df['target_id'].isna().sum())
print('sm w/o categories: ', len(null_categories)) 
```
